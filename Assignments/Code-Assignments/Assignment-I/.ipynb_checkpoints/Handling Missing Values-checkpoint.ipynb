```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('titanic.csv') #load data-set
```

```python
msno.matrix(df)
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
missing_info = pd.DataFrame({
    'MissingCount' : missing_values,
    'MissingPercentage' :(missing_values / len(df)) * 100
})

missing_info = missing_info[missing_info['MissingCount'] > 0].sort_values(by = 'MissingPercentage', ascending=False)
missing_info
```

```python
#Impute missing numeric values with mean
numeric_df = df.select_dtypes(include = 'number')
numeric_df.fillna(numeric_df.mean() , inplace=True)
```

```python
#Deletion
#Remove Rows with Missing Values
numeric_df.dropna(axis=0,inplace=True)
```

```python
#Remove Columns with Missing Values
numeric_df.dropna(axis =1, inplace=True)
```

```python
#comparision between before and after results
numeric_df.describe()
```

```python
numeric_df_imputed = numeric_df.copy()
numeric_df_imputed.fillna(numeric_df.mean(),inplace=True)
```
